## Co-occurence graph of the skills 

```python
import pandas as pd
import pandas_profiling
import numpy as np
from glob import glob
import os
import matplotlib.pyplot as plt
#import graph_tool.all as gt
import seaborn as sns
import matplotlib.style
import matplotlib as mpl
from collections import Counter
import re
import networkx as nx
```

```python
def connect_nodes(list_of_nodes):
    set_of_nodes=set(list_of_nodes)
    nodes=sorted(list(set_of_nodes))
    edges=[]
    if len(nodes)>1:
        for i in range(len(nodes)):
            for j in range(i+1, len(nodes)):
                edges += [(nodes[i], nodes[j])]
    elif len(nodes)==1:
        edges += [(nodes[0], nodes[0])]
    return edges
```

```python
def replace_keywords(data_frame):
    keywords_in_lists = data_frame['allSkills'].dropna().apply(lambda row: row.split(', '))
    keywords = [item for sublist in keywords_in_lists for item in sublist]
    keywords = [item.lower() for item in keywords]
    keywords = [item.replace('networks', 'network') for item in keywords]
    keywords = [item.replace('graphs', 'graph') for item in keywords]
    keywords = [item.replace('scale free', 'scale-free') for item in keywords]
    keywords = [item.replace('small world', 'small-world') for item in keywords]
    keywords = [item.replace('small-world network', 'small-world') for item in keywords]
    keywords = [item.replace('scale-free network', 'scale-free').replace('power laws', 'power-law').replace('power law', 'power-law').replace('power-law distribution', 'power-law') for item in keywords]
    keywords = [item.replace('network theory (graph)', 'network theory') for item in keywords]
    keywords = [item.replace('small-worlds', 'small-world') for item in keywords]
    keywords = [item.replace('peer-to-peer network', 'peer-to-peer') for item in keywords]
    return keywords
```

```python
def keyword_graph(data_frame):
    keywords_in_lists = data_frame['allSkills'].dropna().apply(lambda row: row.split(', '))
    keywords_in_lists = keywords_in_lists.apply(lambda ls: [(item.lower()
                                                             .replace('networks', 'network')
                                                             .replace('graphs', 'graph')
                                                             .replace('scale free', 'scale-free')
                                                             .replace('small world', 'small-world')
                                                             .replace('small-world network', 'small-world')
                                                             .replace('scale-free network', 'scale-free')
                                                             .replace('power laws', 'power-law')
                                                             .replace('power law', 'power-law')
                                                             .replace('power-law distribution', 'power-law')
                                                             .replace('network theory (graph)', 'network theory')
                                                             .replace('small-worlds', 'small-world')
                                                             .replace('peer-to-peer network', 'peer-to-peer')
                                                             .replace('zipf law', "zipf's law")
                                                             .replace("'scale-free' network", 'scale-free network')) for item in ls])
    edges_in_lists = keywords_in_lists.apply(connect_nodes)
    edges_and_sizes = dict(Counter([item for sublist in edges_in_lists.tolist() for item in sublist]))
    keyword_network_df = pd.DataFrame.from_dict(edges_and_sizes, orient='index', columns=['edge_size']).reset_index().rename(index=str, columns={'index':'edge'})
    keyword_network_df['source_node'] = keyword_network_df.edge.apply(lambda row: tuple(row)[0])
    keyword_network_df['target_node'] = keyword_network_df.edge.apply(lambda row: tuple(row)[1])
    keyword_network_df.drop(columns=['edge'], inplace=True)
    return keyword_network_df
```

```python
df = pd.DataFrame(list(full_df["allSkills"]), columns =['allSkills']) 
```

```python
keyword_graph_edges = keyword_graph(df)
```

```python
keyword_graph_edges.sort_values('edge_size', ascending=False).head()
```

```python
keywords_nodes = pd.DataFrame.from_dict(dict(Counter(replace_keywords(df))), orient='index', columns=['frequency']).reset_index().rename(index=str, columns={'index':'keyword'})
```

```python
plotter = graphistry.bind(source='source_node', destination='target_node', node='keyword', edge_weight='edge_size',
                          point_size='frequency').edges(keyword_graph_edges)
# plotter.plot()
plotter.plot(nodes=keywords_nodes)
```
